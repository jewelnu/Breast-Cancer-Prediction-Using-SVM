<h2 align='center' style='color:purple'>Finding best model and hyper parameter tunning using GridSearchCV</h2>

**Load iris flower dataset**

```python
import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
df
```

```python
df['flower']=iris.target
df
```

<h3 style='color:blue'>Approach 1: Use train_test_split and manually tune parameters by trial and error</h3>

```python
x = df.drop('flower', axis=1)
x
```

```python
y = df['flower']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='rbf', C=1, gamma='auto')

model.fit(X_train,y_train)

model.score(X_test, y_test)
```

<h3 style='color:blue'>Approach 2: Use K Fold Cross validation</h3>

**Manually try suppling models with different parameters to cross_val_score function with 5 fold cross validation**

```python
from sklearn.model_selection import cross_val_score
```

```python
sv1=cross_val_score(SVC(kernel='rbf',C=1,gamma='auto'),x, y, cv=5)
sv1
```

```python
sv1.mean()
```

```python
sv2=cross_val_score(SVC(kernel='poly',C=1,gamma='auto'),x, y, cv=5)
sv2
```

```python
sv2.mean()
```

```python
sv3=cross_val_score(SVC(kernel='rbf',C=20,gamma='auto'),x, y, cv=5)
sv3
```

```python
sv3.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
sv3 = cross_val_score(DecisionTreeClassifier(), x, y, cv=5)
sv3
```

```python
sv3.mean()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
sv3 = cross_val_score(GaussianNB(), x, y, cv=5)
sv3
```

```python
sv3.mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
sv4 = cross_val_score(KNeighborsClassifier(n_neighbors=3), x, y, cv=5)
sv4
```

```python
sv4.mean()
```

<h3 style='color:blue'>Use GridSearchCV</h3>

**GridSearchCV does exactly same thing as for loop above but in a single line of code**

```python
from sklearn.model_selection import GridSearchCV
```

```python

clf = GridSearchCV(SVC(gamma='auto'), {
    'C': [1,10,20,30],
    'kernel': ['rbf','linear', 'poly']
}, cv=5)

clf.fit(x, y)

clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
clf.best_params_
```

```python
clf.best_score_
```
