<h3>Breast Cancer prediction Using SVM</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Breast_Cancer_Diagnostic.csv')
df
```

```python
df.info()
```

```python
X = df.drop('diagnosis', axis=1)
X
```

```python
y= df.diagnosis
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

Few steps that you can lead me through now!
* Importing classifier, 
* creating instance, 
* train (fit) the model on training data and 
* do the predictions for test data, right? <br>

Let's do this step by step. Note, SVC (support vector classifier) is in the svm class!

### Importing and  training the Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear',C=30,gamma='auto')
#kernel{â€˜linearâ€™, â€˜polyâ€™, â€˜rbfâ€™, â€˜sigmoidâ€™, â€˜precomputedâ€™} or callable, default=â€™rbfâ€™, c=1,10,20,30
```

```python
# training / fitting the model 
svm_model.fit(X_train,y_train)
```

```python
svm_model.score(X_test, y_test)
```

### Predictions  on own data

```python
svm_model.predict([[1.740, 1.91, 10.12, 585.0, 0.07944, 0.06376, 1.02881, 0.01329, 0.1473, 2.05580]])
```

### Lets go with Grid search algorithm to find parameter for best score
**GridSearchCV does exactly same thing as for loop above but in a single line of code**

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(SVC(gamma='auto'), {
    'C': [1,10,20,30],
    'kernel': ['rbf','linear'] # when we use three parameter then processing time is too high
}, cv=5)
clf.fit(X,y)
clf.cv_results_
```

```python
df_gd = pd.DataFrame(clf.cv_results_)
df_gd
```

```python
df_gd[['param_C','param_kernel','mean_test_score']]
```
